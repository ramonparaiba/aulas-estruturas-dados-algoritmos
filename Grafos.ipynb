# Grafos


```python
#Matriz de adjacência

def criar_matriz_adj(V):
  return [[0 for _ in range(V)] for _ in range(V)]

def adicionar_aresta_matriz(matriz, u, v):
  matriz[u][v] = 1
  matriz[v][u] = 1
```

```python
matriz = criar_matriz_adj(50)
adicionar_aresta_matriz(matriz, 0,1)
adicionar_aresta_matriz(matriz, 0,2)
adicionar_aresta_matriz(matriz, 3,2)
adicionar_aresta_matriz(matriz, 2,2)


for linha in matriz:
  print(linha)
```

```python
#LISTA DE ADJACENCIAS

def criar_lista_adj(V):
  return {i: [] for i in range(V)}

def adicionar_aresta_lista(lista, u, v):
  lista[u].append(v)
  lista[v].append(u) #grafo não dirigido
```

```python
lista = criar_lista_adj(50)
adicionar_aresta_lista(lista, 0, 1)
adicionar_aresta_lista(lista, 0, 2)
print(lista)
```
